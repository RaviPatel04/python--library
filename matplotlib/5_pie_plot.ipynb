```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
e = [0.2,0,0,0]
plt.pie(x,labels = y)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
e = [0.2,0,0,0]
c = ["b","y","g","r"]
plt.pie(x,labels = y,explode = e,colors = c,autopct = "%0.1f%%",shadow = True,labeldistance = 1.12)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
e = [0.2,0,0,0]
c = ["b","y","g","r"]
plt.pie(x,labels = y,explode = e,colors = c,autopct = "%0.1f%%",labeldistance = 1.12,startangle = 90,textprops = {"fontsize":14},radius = 1.5)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
e = [0.2,0,0,0]
c = ["b","y","g","r"]
plt.pie(x,labels = y,explode = e,colors = c,autopct = "%0.1f%%",counterclock = False, wedgeprops = {"linewidth": 5, "edgecolor":"m"},rotatelabels = True)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
plt.pie(x,labels = y)
plt.pie([1],radius = .7,colors = "w")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,1,4,2,5]
plt.plot(x,y)
plt.title("Tata Share",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Price",fontsize = 15)
plt.text(3,4,"High",fontsize = 11,color = "b",style = "italic",bbox = {"facecolor":"r"})
plt.legend(["Growth"],loc  = 2,facecolor = "g",edgecolor = "r",framealpha = .5,shadow = True)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,1,4,2,5]
plt.plot(x,y)
plt.title("Tata Share",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Price",fontsize = 15)
plt.text(3,4,"High",fontsize = 11,color = "b")
plt.annotate("Low",xy = (2,1),xytext = (3,1.5), arrowprops = dict(facecolor = "r",shrink = 10))
plt.show()
```

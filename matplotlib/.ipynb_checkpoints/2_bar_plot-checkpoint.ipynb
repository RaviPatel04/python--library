```python
import matplotlib.pyplot as plt
```

```python
x = ["Python","Java","c","c++","Mojo"]
y = [4.8,4.4,3.7,3.2,2]
plt.xlabel("Language",fontsize = 12)
plt.ylabel("Rating",fontsize = 12)
plt.title("Most like language")
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,width = .5,color = ["r","b","g","y","c"],align = "edge",edgecolor = "k",linewidth = 2,label = "Most like lang.")
plt.legend()
```

```python
plt.bar(x,y,width = .5,color = ["r","b","g","y","c"],align = "edge",edgecolor = "k",linewidth = 2,linestyle = ":",alpha = .5)
```

```python
z = [4.2,3.8,3.3,3,1.5]
plt.bar(x,y,width = .5,color = "r",label = "Most like lang.")
plt.bar(x,z,width = .5,color = "g",label = "Easy to understand")
plt.legend()
plt.show()
```

```python
import numpy as np

width = .25
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.bar(p,y,width,color = "r",label = "Most like lang.")
plt.bar(p1,z,width,color = "g",label = "Easy to understand")

plt.xticks(p+width/2,x,rotation = 30)
plt.legend()
plt.show()
```

```python
plt.barh(p,y,width,color = "r",label = "Most like lang.")
plt.barh(p1,z,width,color = "g",label = "Easy to understand")

plt.yticks(p+width/2,x)
plt.legend()
plt.show()
```

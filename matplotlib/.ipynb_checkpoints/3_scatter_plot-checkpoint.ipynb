```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,6,5,3,1,2]
size = [40,50,60,70,80,90]
plt.scatter(x,y,color = "g",s = size)
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,6,5,3,1,2]
size = [40,50,60,70,80,90]
plt.scatter(x,y,color = "g",s = size,alpha = .5,marker = "*")
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,6,5,3,1,2]
size = [100,200,300,400,500,600]
plt.scatter(x,y,color = "g",s = size,marker = "*",edgecolor = "k",linewidth = 1.5)
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,6,5,3,1,2]
size = [40,50,60,70,80,90]
color = [20,30,40,50,60,70]
plt.scatter(x,y,c = color,s = size,cmap = "hsv") #colormap(cmap) is used plot color using range
plt.colorbar(label = "Colorbar")
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [4,3,2,1]
plt.title("fill_between")
plt.plot(x,y,color = "k")
plt.fill_between(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [4,3,2,1]
plt.plot(x,y,color = "k")
plt.fill_between(x = [2,3],y1 = 2, y2 = 3,color = "r")
plt.show()
```

```python
import numpy as np

x = np.array([1,2,3,4])
y = np.array([4,3,2,1])
plt.plot(x,y,color = "k")
plt.fill_between(x,y,color = "b",where = (x>=2) & (x<=3),alpha = .5)
plt.show()
```

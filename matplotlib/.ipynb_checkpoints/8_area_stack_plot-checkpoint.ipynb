```python
import matplotlib.pyplot as plt
```

```python
x = [0,4,5,6,7]
area = [1,2,5,4,7]
area1 = [2,1,2,1,7]
area2 = [4,2,5,6,8]
plt.stackplot(x,area,area1,area2,labels = ["area","area1","area2"],colors = ["r","y","g"])
plt.legend(loc = 2)
plt.show()
```

```python
x = [0,4,5,6,7]
area = [1,2,5,4,7]
area1 = [2,1,2,1,7]
area2 = [4,2,5,6,8]
plt.stackplot(x,area,area1,area2,baseline = "sym")
plt.show()
```

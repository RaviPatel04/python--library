```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,labels = ["1st"])
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,notch = True)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,labels = ["1st"],patch_artist = True)
plt.show()
```

```python
x = [10,20,30,40,50,100]
plt.boxplot(x,labels = ["1st"],patch_artist = True,whis = 2,sym = "r")
plt.show()
```

```python
x = [10,20,30,40,50]
y = [90,80,70,60,50]
z = [x,y]
plt.boxplot(z,labels = ["1st","2nd"],boxprops = dict(color = "m"),capprops = dict(color = "b"),whiskerprops = dict(color = "r"))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x = np.random.randint(10,80,(50))
x
```

```python
plt.hist(x,color = "r",edgecolor = "k")
plt.title("histogram plot")
plt.xlabel("number")
plt.ylabel("repetition")
plt.show()
```

```python
b = [10,20,30,40,50,60,70,80]
plt.hist(x,color = "r",bins = b,edgecolor = "k")
plt.title("histogram plot")
plt.xlabel("number")
plt.ylabel("repetition")
plt.show()
```

```python
b = [10,20,30,40,50,60,70,80]
plt.hist(x,"auto",(0,100),edgecolor = "b",color = "g")
plt.title("histogram plot")
plt.xlabel("number")
plt.ylabel("repetition")
plt.show()
```

```python
b = [10,20,30,40,50,60,70,80]
plt.hist(x,edgecolor = "b",color = "g",histtype = "step")
plt.title("histogram plot")
plt.xlabel("number")
plt.ylabel("repetition")
plt.show()
```

```python
b = [10,20,30,40,50,60,70,80]
plt.hist(x,edgecolor = "b",color = "g",orientation = "horizontal",rwidth = .8)
plt.title("histogram plot")
plt.ylabel("number")
plt.xlabel("repetition")
plt.show()
```

```python
b = [10,20,30,40,50,60,70,80]
plt.hist(x,color = "r",bins = b,edgecolor = "k",label = "Number")
plt.axvline(45,color = "k")
plt.title("histogram plot")
plt.xlabel("number")
plt.ylabel("repetition")
plt.legend()
plt.grid()
plt.show()
```
